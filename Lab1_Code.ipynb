```python
import pandas as pd
import matplotlib.pyplot as plt

df= pd.read_csv("Stroke.csv")

print(df.head())
```

```python
plt.figure()
df['age'].hist()

plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
```

```python
plt.figure()
df.boxplot('age')


plt.title('Boxplot(Age)')
plt.xlabel('Age')
plt.ylabel('Frequency')
```

```python
print(df.isnull().sum())
missing_rows = df[df.isnull().any(axis=1)]

```

```python
df= df.dropna()
print(df.isnull().sum())
missing_rows = df[df.isnull().any(axis=1)]
display(df)

```

```python
plt.boxplot(df['age'])

age_IQR = df['age'].quantile(0.75) - df['age'].quantile(0.25)
age_low_limit = df['age'].quantile(0.25) - 1.5 * age_IQR
age_upper_limit = df['age'].quantile(0.75) + 1.5 * age_IQR
df[df['age'] > age_upper_limit]
df.drop(df[df['age'] < age_low_limit].index, inplace = True)
df.drop(df[df['age'] > age_upper_limit].index, inplace = True)

plt.boxplot(df['age'])
```

```python
sample_df = df.sample(n=30, random_state=22)
newcolumns_df = sample_df.drop(columns=['work_type', 'Residence_type'])
display(df)
display(newcolumns_df)
```
